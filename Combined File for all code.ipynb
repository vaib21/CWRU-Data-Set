```python
import numpy as np
import pandas as pd
from tftb.processing import WignerVilleDistribution,PseudoWignerVilleDistribution,smoothed_pseudo_wigner_ville
```

```python
data=pd.read_csv('0hp_all_faults.csv')
signal=data.iloc[:,0]
signal=signal.values
# signal=signal.tolist()
# signal=np.array(signal)
signal.shape
```

```python
errors=data.fault.unique()
errors=errors.tolist()
```

# WVD

```python
for error in errors:
    temp=data[data['fault']==error]
    signal=temp.iloc[0:1000:,0].values
    print("The type of fault is ",error)
    obj=WignerVilleDistribution(signal)
    obj.run()
    obj.plot(threshold=0.0, kind='wireframe')
```

# PWVD

```python
for error in errors:
    temp=data[data['fault']==error]
    signal=temp.iloc[0:1000:,0].values
    print("The type of fault is ",error)
    obj=PseudoWignerVilleDistribution(signal)
    obj.run()
    obj.plot()
```

# SPWVD

```python
for error in errors:
    temp=data[data['fault']==error]
    signal=temp.iloc[0:1000:,0].values
    print("The type of fault is ",error)
    obj=smoothed_pseudo_wigner_ville(signal)
    print(obj)
```
